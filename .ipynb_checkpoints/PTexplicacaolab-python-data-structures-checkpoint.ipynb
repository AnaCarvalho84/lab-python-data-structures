# Exercicio: Gestão de Encomendas Clientes


Como parte de um negocio estou a fazer uma loja online que vende vários produtos. Para garantir operações suaves, preciso de desenvolver um programa que faça a gestão dos pedidos dos clientes e do inventário

Segue os seguintes passos para o exercicio:
1. Definir uma lista chamada products que contenham os seguintes items: "t-shirt", "mug", "hat", "book", "keychain".
2. Criar um dicionário vazio com nome de inventory
3. Pedir ao cliente para introduzir a quantidade de produto disponível no inventario. Usa os nomes dos produtos da lista products como keys no dicionario inventory e atribua a respetiva quantidade como valores
4. Cria uma set vazia de nome customer_orders
5. Pede ao cliente para dizer 3 produtos que quer encomendar
6. Print os produtos no customer_orders set
7. Calcula as seguintes estatisticas:
    a. Total dos Produtos Ordered: O total do numero dos produtos no customer_orders set
    b. Unique Prodcuts Ordered: O numero dos unicos produtos orders na customer_orders set
    c. Percentagem dos Unique Products Ordered: A % dos unique productos ordered em comparação com o total dos produtos disponiveis
    Coloca estas estatisticas numa tupla com nome order_status
 8. Print a order statistics no seguinte formato:
  Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Unique Products Ordered: <unique_products_ordered>
   Percentage of Unique Products Ordered: <percentage_ordered>% 
  
 9. Uptade o inventario subtraindo 1 da quantidade de cada produto. Modifica o inventory dict 

10. Print o uptdate inventory, mostrando a quantidade de cada produto em linhas separadas

Resolve o exercicio usando Listas, Dicionarios, Sets e input/output 
 

```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {} #empty dictionary

for product in products:
    quantity = int(input(f"Enter the quantity of {product} available: ")) #int so set a number
    inventory[product] = quantity 

print("The products avaliable on inventory are: ", inventory)            
print("\n")    
# Create an empty set called customer_orders
customer_orders = set()
inventory_sum = sum (inventory.values())  #number of products on my inventory


# Ask the user to input the name of three products that a customer wants to order. Add each product name to the customer_orders set.
while products == ["t-shirt", "mug", "hat", "book", "keychain"] and len(customer_orders)<3:
    product_name = input('Enter the name of a product: ')
    if product_name in products:
        customer_orders.add(product_name)
    else:
        print("Invalid product name. Try again")

print("The customer orders are: ", customer_orders)


customer_number_order = {}
for item in customer_orders:
    while True:
        number = int(input(f"Enter the quantity of {item} you want: "))
        if number <= inventory[item]:
            customer_number_order[item] = number
            inventory[item] -= number 
            break 
        else:
            print(f"Insufficient inventory for {item}. Please choose a lower quantity.")

print("The Customer number order are: ", inventory)
 
print("The total products on inventory are: ", inventory_sum)
print("\n" + "-" * 30) 
print("Order Statistics")
total_product_order = sum (customer_number_order.values()) 
unique_products_ordered = len(customer_orders)
percentage_ordered = unique_products_ordered / inventory_sum *100
print('The total of products ordered ', total_product_order)
print('Unique products ordered: ', unique_products_ordered)
print('The percentage of unique products ordered are:',percentage_ordered,"%")

print("\n" + "-" * 30)  


for key in inventory:
    inventory[key] -=1 

for key, item in inventory.items():
    print("For the product:",key, "the quantity in inventory are:",item)

print()
```
